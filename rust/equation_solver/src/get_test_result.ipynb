```python
import sympy as sp
x=sp.Symbol('x')
eq=sp.Eq(x**3 - 2.0 * x**2 + 3.0 * x - 5.0,0)
r=sp.solve(eq,x)
print(r)
```

```python
import sympy as sp
z=sp.Symbol('z')
eq=sp.Eq(z**3 - 2.0 * z**2 + 4.0/3.0 * z - 8/27,0)
r=sp.solve(eq,z)
print(r)
r=sp.roots(eq,z)
print(r)
```
